```python
import pandas
data = pandas.read_csv( 'USA_compulsory_sterilization_totals_years.csv')
X = data.columns
X = list(X[5:-2])
# labels = data.values[:,0]
# Y = list( data.values[:,5:-2].tolist() )
```

```python
# Select by area
data_northeast   = data[data['region'] == 'northeast']
labels_northeast = data_northeast.values[:,0]
Y_northeast = list( data_northeast.values[:,5:-2].tolist() )

data_west   = data[data['region'] == 'west']
labels_west = data_west.values[:,0]
Y_west = list( data_west.values[:,5:-2].tolist() )

data_midwest   = data[data['region'] == 'midwest']
labels_midwest = data_midwest.values[:,0]
Y_midwest = list( data_midwest.values[:,5:-2].tolist() )

data_southeast   = data[data['region'] == 'southeast']
labels_southeast = data_southeast.values[:,0]
Y_southeast = list( data_southeast.values[:,5:-2].tolist() )
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# pal = sns.color_palette("Reds_r", n_colors=9)
fig = plt.figure(num=None, figsize=(26, 16), dpi=80, facecolor='w', edgecolor='k')

fig.add_subplot(2,2,1)
a = plt.stackplot(X, sorted(Y_northeast, key=sum, reverse=True ), 
              labels=labels_northeast, colors=sns.color_palette("Greys_r", n_colors=len(labels_northeast)+2))
plt.legend(loc='upper right')
plt.axis([0, 20, 0, 850])
# plt.ylim(900)

# plt.show()

plt.subplot(2,2,2)
b = plt.stackplot(X,sorted(Y_west, key=sum, reverse=True ), 
              labels=labels_west, colors=sns.color_palette("Greys_r", n_colors=len(labels_west)+2))
plt.legend(loc='upper right')
plt.axis([0, 20, 0, 850])
# plt.show()

plt.subplot(2,2,3)
c = plt.stackplot(X,sorted(Y_midwest, key=sum, reverse=True ), 
              labels=labels_midwest, colors=sns.color_palette("Greys_r", n_colors=len(labels_midwest)+2))
plt.legend(loc='upper right')
plt.axis([0, 20, 0, 850])
# plt.show()

plt.subplot(2,2,4)
d = plt.stackplot(X,sorted(Y_southeast, key=sum, reverse=True ), 
              labels=labels_southeast, colors=sns.color_palette("Greys_r", n_colors=len(labels_southeast)+2))
plt.legend(loc='upper right')
plt.axis([0, 20, 0, 850])
plt.show()


```
